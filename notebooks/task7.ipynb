#### Построить график функции
$$
\begin{align}
Z = 5\times y^2-x^2  \\
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-3, 3, 0.1)
y = np.arange(-3, 3, 0.1)
z = np.empty((x.size, y.size))
for i, el_x in np.ndenumerate(x):
    for j, el_y in np.ndenumerate(y):
        z[i][j] = 5 * (el_y) ** 2 - (el_x) ** 2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')
```
